```python
import pandas as pd
from geopy import distance
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 500)

```

```python
sacoleiras = pd.read_csv('data/sacoleiras.csv', sep=';')
pedidos = pd.read_csv('data/pedidos.csv', sep=';')
```

```python
id_vendedor_buscado = 23
pedidos_promotor = pedidos[pedidos['id_vendedor']==id_vendedor_buscado]
pedidos_promotores = pedidos[pedidos['id_vendedor']!=id_vendedor_buscado]

pedidos_promotor = pedidos_promotor[(pedidos_promotor['status'] == "DIGITADO") | (pedidos_promotor['status'] == "TRANSFERIDO")]
pedidos_promotores =  pedidos_promotores[(pedidos_promotores['status'] == "DIGITADO")& (pedidos_promotores['ano_cobranca'] == 2024)]
```

```python
cidades_promotor = pedidos_promotor[["cidade", "latitude", "longitude", "descricao_vendedor", "descricao_praca"]].drop_duplicates().dropna()
cidades_promotores = pedidos_promotores[["cidade", "latitude", "longitude", "descricao_vendedor", "descricao_praca"]].drop_duplicates().dropna()
```

```python
resultados = []
for _, up_row in cidades_promotor.iterrows():
    lat_long_busca = (up_row['latitude'], up_row['longitude'])
    for _, row in cidades_promotores.iterrows():
        lat_long_teste = (row['latitude'], row['longitude'])
        distancia = distance.distance(lat_long_busca, lat_long_teste).km
        if distancia < 30:
            # Adicionar os resultados Ã  lista
            resultados.append({
                'cidade_buscada': up_row['cidade'],
                'vendedor': row['descricao_vendedor'],
                'praca': row['descricao_praca']
            })
lista_coincidentes = pd.DataFrame(resultados)
```

```python
lista_coincidentes.pivot_table(index=['cidade_buscada', 'vendedor', 'praca'], aggfunc='first')
```
